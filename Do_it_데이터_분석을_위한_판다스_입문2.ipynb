## 12-1 datetime 오브젝트
datetime 라이브러리: 날씨와 시간을 처리하는 등의 다양한 기능을 제공

날짜를 처리하는 date 오브젝트, 시간을 처리하는 time 오브젝트, 날짜와 시간을 모두 처리하는 datetime 오브젝트

```python
from datetime import datetime
now1=datetime.now() # 현재시간
print(now1)

now2=datetime.today()
print(now2)
```

```python
# datetime 오브젝트 생성

t1=datetime.now()
t2=datetime(1970,1,1)
t3=datetime(1970,12,12,13,24,34)

print(t1)
print(t2)
print(t3)
```

```python
# 시간 계산 가능
diff1=t1-t2
print(diff1)
print(type(diff1))

diff2=t2-t1
print(diff2)
print(type(diff2))
```
