### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기 **df.DataFrame()**

-DataFrame은 리스트, 넘파이 ndarray와 달리 column명을 가짐

-리스트, 넘파이 ndarray를 DataFrame으로 변환 시, column명을 지정(지정하지 않으면 자동으로 column명을 할당)

-2차원 이하의 데이터들만 DataFrame으로 변환 가능

```python
# 1차원 형태의 데이터를 기반으로 DataFrame을 생성하므로 column명은 1개만 필요

import numpy as np

col_name1=['col1']
list1=[1,2,3]
array1=np.array(list1)
print('array1 shape:',array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1=pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame:',df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1=pd.DataFrame(array1,columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:',df_array1)
```

```python
# 2행 3열 형태의 리스트와 ndarray를 기반으로 DataFrame을 생성하므로 3개의 column명이 필요함
col_name2=['col1','col2','col3']

# 2행x3열의 리스트와 ndarray를 생성한 뒤 이를 DataFrame으로 변환
list2=[[1,2,3],
       [11,12,13]]
array2=np.array(list2)
print('array2 shape:',array2.shape)

df_list2=pd.DataFrame(list2,columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n',df_list2,'\n')

df_array2=pd.DataFrame(array2,columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n',df_array2)
```

```python
# 딕셔너리를 DataFame으로 변환 시, 딕셔너리의 Key는 column명으로, Value는 column데이터로 변환
# Key는 문자열, Value는 리스트, ndarray 형태로 딕셔너리를 구성

# Key는 문자열 column명으로 매핑, Value는 리스트 형(또는 ndarray) column 데이터로 매핑
dict={'col1':[1,11],'col2':[2,22],'col3':[3,33]}
print(dict,'\n')

df_dict=pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n',df_dict)
```

### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기 

-DataFrame -> ndarray: **df.values**

-DataFrame -> list: **df.values.tolist()**

```python
# DataFrame을 list로 변환
list3=df_dict.values.tolist()
print('df_dict.values.tolist() 타입:',type(list3))
print(list3)
```

```python
# DataFrame을 딕셔너리로 변환
dict3=df_dict.to_dict('list') # 인자로 'list'를 입력 시, 딕셔너리의 값(value)이 리스트형으로 반환됨
print('df_dict.to_dict() 타입:',type(dict3))
print(dict3)
```

### apply lambda 식으로 데이터 가공

lambda x: 반환 값 if 조건 else 반환값 (else if 는 지원X)

```python
def get_square(a):
  return a**2
print('3의 제곱은',get_square(3))
```

```python
# lambda : 함수의 선언과 함수 내의 처리를 한 줄의 식으로 쉽게 변환하는 식
lambda_square=lambda x: x**2
print('3의 제곱은',lambda_square(3))
```

```python
# lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우 map()함수 이용
a=[1,2,3]
squares=map(lambda x:x**2,a)
list(squares)
```

```python
# Name 열의 문자열 개수를 Name_len에 저장
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x:len(x))
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x:'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# else 가 많이 나오는 경우는 사용이 어려우므로 별도의 함수 생성

# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat=''
  if age<=5: cat = 'Baby'
  elif age<=12: cat='Child'
  elif age<=18: cat='Teenager'
  elif age<=25: cat='Student'
  elif age<=35: cat='Young Adult'
  elif age<=60: cat='Adult'
  else: cat='Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category()함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age'열 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
```
